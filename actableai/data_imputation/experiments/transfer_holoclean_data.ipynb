```python
DATA_TYPE = "adult"

# MY_FIX_CSV = "./data/adult/adult.csv"
CLEAN_ORIGIN_CSV = f"./data/{DATA_TYPE}/{DATA_TYPE}_clean_original.csv"
CLEAN_CSV = f"./data/{DATA_TYPE}/{DATA_TYPE}.csv"

# with open(MY_FIX_CSV, "r") as f:
#     head = f.readline().strip()

# columns = head.split(",")
# columns
```

```python
from collections import defaultdict
import pandas as pd


data = defaultdict(list)

with open(CLEAN_ORIGIN_CSV, "r") as f:
    l_c = -1
    for l in f:
        l_c += 1
        if l_c == 0:
            continue
        line = l.strip()
        tid, attribute, val = line.split(",")
        data[attribute].append(val)
        
full_csv = pd.DataFrame.from_dict(data)
# full_csv = full_csv[columns]

full_csv.head()
```

```python
full_csv.to_csv(CLEAN_CSV, index=False)
```
