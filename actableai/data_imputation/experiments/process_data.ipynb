# Remove some data

```python
import random
from collections import namedtuple, defaultdict


DropInfo = namedtuple("DropInfo", ["col", "index"])


def drop_random_cells(full_csv, drop_percent=0.1):
    df = pd.read_csv(full_csv)

    drop_info = []
    drop_count_total_expect = int(len(df.columns) * len(df) * drop_percent)

    for _ in range(drop_count_total_expect):
        c = random.choice(df.columns)
        i = random.randint(0, len(df))
        df.at[i, c] = None
        drop_info.append(DropInfo(c, i))

    return df, drop_info


def write_drop_info(drop_info, readme_path):
    grouped_drop_info = defaultdict(list)
    for info in drop_info:
        grouped_drop_info[info.col].append(info.index)

    with open(readme_path, "w") as f:
        f.write("col,index\n")
        for col in grouped_drop_info:
            f.write(f"{col},{sorted(grouped_drop_info[col])}\n")


processed_data, drop_info = drop_random_cells(PATH)
processed_data.to_csv(DROP_DATA_PATH, index=False)
write_drop_info(drop_info, DATA_DROP_INFO_PATH)
```
