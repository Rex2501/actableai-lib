```python
df = pd.read_csv("/mnt/c/Users/trung/Downloads/apartments.csv")
re = AAIClusteringTask().run(df, num_clusters="auto", alpha_k=0.02, explain_precision_threshold=0.7)
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
for cluster, c in zip(re["data"], ["r", "g", "b", "c", "m", "y"]):
    pd.DataFrame([d["train"] for d in cluster["value"]]).plot.scatter(x="x", y="y", ax=ax, c=c, s=1)
#     ax.scatter([cluster["projected_value"][0]], [cluster["projected_value"][1]], c=c, marker="x")
    ax.scatter([cluster["projected_nearest_point"][0]], [cluster["projected_nearest_point"][1]], c=c, marker="x", s=400)
    print(cluster["explanation"])
    print(cluster["projected_nearest_point"])
```
